```python
import ipywidgets as widgets
from IPython.display import display

# Creamos una matriz de botones para representar el tablero del juego
buttons = [widgets.Button(description='', layout=widgets.Layout(width='100px', height='100px')) for _ in range(9)]

# Creamos una lista para representar el estado actual del tablero
board = [''] * 9

# Función para manejar el clic en un botón del tablero
def on_button_click(button, index):
    global board
    if button.disabled or board[index] != '':
        return
    button.description = 'X'
    board[index] = 'X'
    if check_winner('X'):
        print("¡Felicidades! ¡Has ganado!")
        disable_all_buttons()
        return
    if '' not in board:
        print("¡Es un empate!")
        disable_all_buttons()
        return
    computer_move()
    if check_winner('O'):
        print("¡El ordenador ha ganado!")
        disable_all_buttons()
        return

# Función para que el ordenador haga su movimiento
def computer_move():
    global board
    best_score = -float('inf')
    best_move = None
    for i, val in enumerate(board):
        if val == '':
            board[i] = 'O'
            score = minimax(board, 0, False)
            board[i] = ''
            if score > best_score:
                best_score = score
                best_move = i
    buttons[best_move].description = 'O'
    board[best_move] = 'O'

# Función para deshabilitar todos los botones del tablero
def disable_all_buttons():
    for button in buttons:
        button.disabled = True

# Función para verificar si un jugador ha ganado
def check_winner(player):
    winning_combinations = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]
    for combo in winning_combinations:
        if all(board[i] == player for i in combo):
            return True
    return False

# Función para el algoritmo min-max
def minimax(board, depth, is_maximizing):
    if check_winner('X'):
        return -10 + depth
    elif check_winner('O'):
        return 10 - depth
    elif '' not in board:
        return 0
    if is_maximizing:
        max_score = -float('inf')
        for i, val in enumerate(board):
            if val == '':
                board[i] = 'O'
                score = minimax(board, depth + 1, False)
                board[i] = ''
                max_score = max(max_score, score)
        return max_score
    else:
        min_score = float('inf')
        for i, val in enumerate(board):
            if val == '':
                board[i] = 'X'
                score = minimax(board, depth + 1, True)
                board[i] = ''
                min_score = min(min_score, score)
        return min_score

# Asignamos la función on_button_click a cada botón del tablero
for i, button in enumerate(buttons):
    button.on_click(lambda _, index=i: on_button_click(buttons[index], index))

# Mostramos los botones del tablero
display(widgets.GridBox(buttons, layout=widgets.Layout(grid_template_columns="repeat(3, 100px)")))
```
